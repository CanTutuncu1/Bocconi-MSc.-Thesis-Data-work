```python
import os
import json
import dataclasses
from pyUSPTO import PTABTrialsClient

API_KEY = os.environ.get("USPTO_ODP_API_KEY")
if not API_KEY:
    raise RuntimeError("Missing USPTO_ODP_API_KEY environment variable.")

client = PTABTrialsClient(api_key=API_KEY)

# Trial types in PTAB trials data
TRIAL_TYPES = ["IPR", "PGR", "CBM", "DER"]

def to_dict(obj):
    """Convert pyUSPTO objects to plain dicts for JSON output."""
    try:
        return dataclasses.asdict(obj)
    except Exception:
        return obj.__dict__ if hasattr(obj, "__dict__") else {"value": str(obj)}

all_rows = []

for t in TRIAL_TYPES:
    print(f"Downloading proceedings for {t} ...")
    for rec in client.paginate_proceedings(trial_type_code_q=t, limit=100):
        all_rows.append(to_dict(rec))

print("Total proceedings rows:", len(all_rows))

out_path = "trial_proceedings.jsonl"
with open(out_path, "w", encoding="utf-8") as f:
    for row in all_rows:
        f.write(json.dumps(row, default=str) + "\n")

print("Wrote:", out_path)
```
